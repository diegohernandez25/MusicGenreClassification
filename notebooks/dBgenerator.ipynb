```
files=["Pickled_Data/Final/WF/sp_x_train.pickle",
      "Pickled_Data/Final/WF/sp_x_valid.pickle",
      "Pickled_Data/Final/WF/sp_x_test.pickle"]

pickles=["Pickled_Data/Final/DB/np_db_x_train1.pickle",
         "Pickled_Data/Final/DB/np_db_x_valid1.pickle",
         "Pickled_Data/Final/DB/np_db_x_test1.pickle"]

for filename,pickelname in zip(files,pickles):
  print(filename)
  f=open(filename,"rb")
  for it in range(0,3):
    print("it:",it)
    X=pickle.load(f)
    Xdb=[]
    for i,spectrogram in enumerate(X,1):
      db=librosa.power_to_db(spectrogram,ref=np.max)  
      Xdb.append(db)

    pickle_out = open(pickelname,"ab")
    pickle.dump(np.array(Xdb),pickle_out)
    pickle_out.close()
    
  print()
  f.close()


files=["Pickled_Data/Final/SP/sp_x_train.pickle",
      "Pickled_Data/Final/SP/sp_x_valid.pickle",
      "Pickled_Data/Final/SP/sp_x_test.pickle"]

pickles=["Pickled_Data/Final/DB/np_db_x_train2.pickle",
         "Pickled_Data/Final/DB/np_db_x_valid2.pickle",
         "Pickled_Data/Final/DB/np_db_x_test2.pickle"]

for filename,pickelname in zip(files,pickles):
  print(filename)
  f=open(filename,"rb")
  for it in range(0,5):
    print("it:",it)
    X=pickle.load(f)
    Xdb=[]
    for i,spectrogram in enumerate(X,1):
      db=librosa.power_to_db(spectrogram,ref=np.max)  
      Xdb.append(db)

    pickle_out = open(pickelname,"ab")
    pickle.dump(np.array(Xdb),pickle_out)
    pickle_out.close()

  print()
  f.close()
```
